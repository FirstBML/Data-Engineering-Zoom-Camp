```python
import pandas as pd
url_trip="https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet"
df_trip = pd.read_parquet(url_trip, engine='fastparquet')
df_trip.head(20)
```

```python
df_zone=pd.read_csv("https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv")
df_zone.head(20)
```

```python

november_trips = df_trip[
    (df_trip['lpep_pickup_datetime'] >= '2025-11-01') & 
    (df_trip['lpep_pickup_datetime'] < '2025-12-01')
]

# Count trips with distance <= 1 mile
short_trips = november_trips[november_trips['trip_distance'] <= 1]
count = len(short_trips)

print(f"Trips with distance ≤ 1 mile: {count}")
```

```python
# Filter for November 18th, 2025 pickups
nov18_trips = df_trip[
    (df_trip['lpep_pickup_datetime'] >= '2025-11-18') & 
    (df_trip['lpep_pickup_datetime'] < '2025-11-19')
]

# Join with zone lookup to get pickup zone names
nov18_with_zones = nov18_trips.merge(
    df_zone[['LocationID', 'Zone']],
    left_on='PULocationID',  # Pickup LocationID column in trips
    right_on='LocationID',
    how='left'
)

# Sum total_amount by pickup zone
zone_totals = nov18_with_zones.groupby('Zone')['total_amount'].sum().reset_index()

# Find the zone with the highest total
top_zone = zone_totals.loc[zone_totals['total_amount'].idxmax()]

print(f"Pickup Zone: {top_zone['Zone']}")
print(f"Total Amount: ${top_zone['total_amount']:,.2f}")
```

```python
from tqdm.auto import tqdm
```

```python
df_trip = pd.read_parquet(url_trip)
```

```python
from tqdm.auto import tqdm

chunk_size = 10000

for start in tqdm(range(0, len(df_trip), chunk_size)):
    end = start + chunk_size

    df_chunk = df_trip.iloc[start:end]

    df_chunk.to_sql(
        name='green_taxi_data',
        con=engine,
        if_exists='replace',
        index=False
    )
```

```python
zone_chunk_size = 100

for start in tqdm(range(0, len(df_zone), zone_chunk_size)):
    end = start + zone_chunk_size

    df_chunk_z = df_zone.iloc[start:end]

    df_chunk_z.to_sql(
        name='taxi_zone_lookup',
        con=engine,
        if_exists='append' if start > 0 else 'replace',
        index=False
    )
```
